```python
# Internal imports
from deep_learning_land_use_classification import config as config
from deep_learning_land_use_classification.dataset import get_single_label_data

# External imports
from PIL import Image
import matplotlib.pyplot as plt
```

```python
train_df, test_df, class_names, num_classes = get_single_label_data()
```

### Plot images

```python
n = 9
sample_df = train_df.sample(n=n, random_state=42)

fig, axes = plt.subplots(3, 3, figsize=(12, 12))
axes = axes.ravel()

for ax, (_, row) in zip(axes, sample_df.iterrows()):
    img = Image.open(row["path"])
    ax.imshow(img)
    ax.set_title(row["label"])
    ax.axis("off")

plt.tight_layout()
plt.show()
```
